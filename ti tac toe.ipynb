```python
# Initialize a 3x3 board with empty strings
list1 = [['', '', ''], ['', '', ''], ['', '', '']]

# Function to display the current state of the board
def display_current():
    for i, row in enumerate(list1):
        # Join the elements of the row with '|' and ensure each cell is of width 3
        print('|'.join([cell.center(3) for cell in row]))
        # Print a separator after each row except the last one
        if i < 2:
            print('-' * 9)

# Function to get a valid user choice
def user_choice():
    choice = 'wrong'
    acceptable_range = range(1, 10)
    within_range = False

    # Loop until a valid choice is made
    while choice.isdigit() == False or within_range == False:
        choice = input("Enter number between (1-9): ")

        # Check if the input is a digit
        if choice.isdigit() == False:
            print("Sorry, not a digit")
        
        # Check if the digit is within the acceptable range
        if choice.isdigit() == True:
            if int(choice) in acceptable_range:
                within_range = True
            else:
                print("Sorry, you are out of range")
                within_range = False

    # Return the valid choice as an integer
    return int(choice)

# Function to update the board with the player's move
def update_board(choice, player):
    # Mapping of user choices to board positions
    mapping = {
        1: (0, 0), 2: (0, 1), 3: (0, 2),
        4: (1, 0), 5: (1, 1), 6: (1, 2),
        7: (2, 0), 8: (2, 1), 9: (2, 2)
    }
    row, col = mapping[choice]
    # Update the board if the position is empty
    if list1[row][col] == '':
        list1[row][col] = player
    else:
        print("Position already taken, choose another one.")
        return False
    return True

# Function to check if the player has won
def check_win(player):
    # Check rows for a win
    for row in list1:
        if row == [player, player, player]:
            return True

    # Check columns for a win
    for col in range(3):
        if list1[0][col] == player and list1[1][col] == player and list1[2][col] == player:
            return True

    # Check diagonals for a win
    if list1[0][0] == player and list1[1][1] == player and list1[2][2] == player:
        return True
    if list1[0][2] == player and list1[1][1] == player and list1[2][0] == player:
        return True

    return False

# Main function to run the game
def main():
    current_player = 'X'
    while True:
        display_current()  # Display the current board
        choice = user_choice()  # Get user choice
        if update_board(choice, current_player):  # Update the board if the move is valid
            if check_win(current_player):  # Check if the current player has won
                display_current()  # Display the final board
                print(f"Player {current_player} wins!")
                break
            # Switch player
            current_player = 'O' if current_player == 'X' else 'X'

# Run the game
main()
```
